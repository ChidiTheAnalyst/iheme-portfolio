**Global Sports Footwear Strategy Analysis** (2020-2025)
**Objective:** To analyze sales performance, customer behavior, and pricing sensitivity for major footwear brands (Nike, Adidas, New Balance, etc.) and forecast 2026 market trends in the UK and USA.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np

df = pd.read_csv('/content/global_sports_footwear_sales_2018_2026.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
df_hist = df[df['order_date'].dt.year <= 2025].copy()
df_hist.dropna(inplace=True)

print(f"Data Loaded. Historical Range: {df_hist['order_date'].min()} to {df_hist['order_date'].max()}")
```

Sales Performance & Market Dominance (2025)
**Key Findings:**
*   **Volume Leader:** New Balance dominated 2025 volume with **1,510 units sold**.

**Gender Trends:** In 2025, Unisex (351k USD) and Men's (345k USD) shoes outperformed Women's (326k USD), showing a shift toward gender-neutral and male-centric demand.
 **Category Kings:**
    ***Nike** leads in Basketball (42k USD).
    *   **New Balance** leads in Gym and Training.
    *   **Adidas** leads in Lifestyle and Running.



```python
# Revenue & Units 2025
df_2025 = df_hist[df_hist['order_date'].dt.year == 2025].copy()
brand_perf_2025 = df_2025.groupby('brand')['revenue_usd'].sum().sort_values(ascending=False)
brand_units_2025 = df_2025.groupby('brand')['units_sold'].sum().sort_values(ascending=False)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
sns.barplot(x=brand_perf_2025.index, y=brand_perf_2025.values, hue=brand_perf_2025.index, palette='viridis', ax=ax1, legend=False)
ax1.set_title('Revenue by Brand (2025)')
sns.barplot(x=brand_units_2025.index, y=brand_units_2025.values, hue=brand_units_2025.index, palette='magma', ax=ax2, legend=False)
ax2.set_title('Units Sold by Brand (2025)')
plt.show()

# Category Dominance
brand_cat = df_2025.groupby(['category', 'brand'])['revenue_usd'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=brand_cat, x='category', y='revenue_usd', hue='brand', palette='deep')
plt.title('Brand Dominance per Category (2025)')
plt.show()
```

Regional Geography & Customer Profiles (UK vs USA)
**Key Findings:**
*   **Income Impact:** High-income customers drive the UK market **(64k USD)**, whereas medium income  **(63k USD)**.
*   **Customer Satisfaction:** UK customers are more satisfied (4.0 stars) compared to USA customers (3.9 stars).
*   **Global Preference:** The UAE and USA show the strongest preference for **Adidas (35k USD)** and **New Balance (35k USD)** respectively, while Nike maintains strongholds in the UK and Pakistan.

```python
# Income & Ratings
df_2025_regions = df_hist[(df_hist['order_date'].dt.year == 2025) & (df_hist['country'].isin(['UK', 'USA']))].copy()
income_grouped = df_2025_regions.groupby(['country', 'customer_income_level'])['revenue_usd'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=income_grouped, x='country', y='revenue_usd', hue='customer_income_level', palette='viridis')
plt.title('Income Level Impact: UK vs USA (2025)')
plt.show()

# Global Heatmap
brand_pref = df_2025.groupby(['country', 'brand'])['revenue_usd'].sum().reset_index()
brand_pivot = brand_pref.pivot(index='country', columns='brand', values='revenue_usd')
plt.figure(figsize=(12, 7))
sns.heatmap(brand_pivot, annot=True, fmt=",.0f", cmap="YlGnBu")
plt.title('Global Brand Preference Heatmap (2025)')
plt.show()
```

The Psychology of the Discount
**Key Findings:**
*   **The USA Strategy:** Full-price sales generate the highest revenue (40k USD). Discounting in the USA leads to a significant drop in profitability. It can be deduced that most customers in the US are brand loyalist and they are willing to pay as much as sport shoes cost as long as  they are guaranteed to get 'perceived value' for their money.
*   **The UK Strategy:** A small 5% discount is the "Sweet Spot" for revenue (37k USD), while 20% moves the most volume but hurts total profit.
*   **Sensitivity:** Reebok and Adidas are Discount Sensitive in the USA, while brands like Nike and New Balance are "Resistant" meaning they should stay at full price.



```python
# Discount vs Revenue
df_deep_dive = df_hist[(df_hist['order_date'].dt.year == 2025) & (df_hist['country'].isin(['UK', 'USA']))].copy()
rev_by_disc = df_deep_dive.groupby(['country', 'discount_percent'])['revenue_usd'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=rev_by_disc, x='discount_percent', y='revenue_usd', hue='country', palette='viridis')
plt.title('Total Revenue vs Discount % (UK vs USA 2025)')
plt.show()

# Sensitivity Heatmaps
def calc_sens(group): return group['discount_percent'].corr(group['units_sold']) if len(group) > 5 else 0
income_sens = df_deep_dive.groupby(['country', 'customer_income_level'], group_keys=False).apply(calc_sens).unstack()
plt.figure(figsize=(8, 5))
sns.heatmap(income_sens, annot=True, cmap='RdYlGn_r', center=0)
plt.title('Discount Sensitivity by Income Level')
plt.show()
```

Predictive Analysis: The 2026 Outlook
**Key Findings:**
*   **USA Forecast:** **Adidas** is set to lead with **(38,280 USD)** in 2026, outpacing Nike due to stronger growth momentum.
*   **UK Forecast:** **Nike** remains the dominant force, predicted to lead with **(38,561 USD)**.
*   **10% Discount Simulation:** Simulation proves a blanket 10% discount in H1 2026 would be **loss-making** for both brands **(Nike: -9.6k USD loss, Adidas: -855 USD loss)**.


```python
# REVENUE FORECAST 2026
brands = ['Nike', 'Adidas']
countries = ['UK', 'USA']

# data training (2024-2025)
df_train = df_hist[(df_hist['order_date'].dt.year.isin([2024, 2025])) &
                   (df_hist['brand'].isin(brands)) &
                   (df_hist['country'].isin(countries))].copy()

df_train['month_year'] = df_train['order_date'].dt.to_period('M')
monthly_sales = df_train.groupby(['country', 'brand', 'month_year'])['revenue_usd'].sum().reset_index()
monthly_sales['time_index'] = (monthly_sales['month_year'].dt.year - 2024) * 12 + monthly_sales['month_year'].dt.month

results_2026 = []
plt.figure(figsize=(15, 6))
colors = {'Nike': '#1f77b4', 'Adidas': '#d62728'}
styles = {'UK': '-', 'USA': '--'}

for country in countries:
    for brand in brands:
        data = monthly_sales[(monthly_sales['country'] == country) & (monthly_sales['brand'] == brand)]
        if not data.empty:
            model = LinearRegression()
            model.fit(data[['time_index']].values, data['revenue_usd'].values)


            future = np.array(range(25, 37)).reshape(-1, 1)
            pred_2026 = model.predict(future)
            results_2026.append({'Country': country, 'Brand': brand, 'Forecast_2026': pred_2026.sum()})

            # Visualization
            full_x = np.append(data['time_index'].values, future)
            full_y = model.predict(full_x.reshape(-1, 1))
            plt.plot(full_x, full_y, label=f"{brand} {country} (Forecast)", color=colors[brand], linestyle=styles[country], linewidth=2.5)
            plt.scatter(data['time_index'], data['revenue_usd'], color=colors[brand], alpha=0.3)

plt.axvline(x=24.5, color='black', linestyle=':', label='Start of 2026')
plt.title('Business Forecasting: 2026 Revenue Trends (Nike vs Adidas)', fontsize=16)
plt.xlabel('Month Index (Month 25 = Jan 2026)')
plt.ylabel('Monthly Revenue (USD)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.2)
plt.show()

# 10% DISCOUNT SIMULATION
df_sim_data = df_hist[(df_hist['order_date'].dt.year.isin([2024, 2025])) & (df_hist['brand'].isin(brands))].copy()
lift_stats = df_sim_data.groupby(['brand', 'discount_percent'])['units_sold'].mean().unstack()

nike_boost = lift_stats.loc['Nike', 10] / lift_stats.loc['Nike', 0]
adidas_boost = lift_stats.loc['Adidas', 10] / lift_stats.loc['Adidas', 0]

base_rev_2025 = df_sim_data[df_sim_data['order_date'].dt.year == 2025].groupby('brand')['revenue_usd'].sum() / 12
six_month_baseline = base_rev_2025 * 6

sim_results = []
for brand in brands:
    boost = nike_boost if brand == 'Nike' else adidas_boost
    current_rev = six_month_baseline[brand]
    simulated_rev = (current_rev * boost) * 0.90
    sim_results.append({'Brand': brand, 'Scenario': 'Baseline', 'Revenue': current_rev})
    sim_results.append({'Brand': brand, 'Scenario': 'Simulated (10% Disc)', 'Revenue': simulated_rev})

sim_df = pd.DataFrame(sim_results)

plt.figure(figsize=(12, 6))
ax = sns.barplot(data=sim_df, x='Brand', y='Revenue', hue='Scenario', palette='coolwarm')
plt.title('Simulation: Impact of a 10% Discount on H1 2026 Revenue', fontsize=16)
for p in ax.patches:
    if p.get_height() > 0:
        ax.annotate(f'${p.get_height():,.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontweight='bold')
plt.show()

forecast_pivot = pd.DataFrame(results_2026).pivot(index='Country', columns='Brand', values='Forecast_2026')
print("\n--- 2026 REVENUE FORECAST ---")
print(forecast_pivot)
```

#  **Strategic Recommendations**

Based on the multi-dimensional analysis of historical sales (2020-2025) and predictive modeling for 2026, the following strategic actions are recommended:

### **1. The Pricing & Profitability Strategy**
*   **USA "Full-Price" Mandate:** The analysis proved that the USA is not a price-sensitive market. High-volume sales occur at 0% discount (40,340 USD revenue). **Recommendation:** Maintain full pricing for New Arrivals in the US. Discounting by 20% in this region nearly halves revenue without driving significant extra volume.
*   **The UK "5% idea":** While 20% discounts move the most units in the UK, the 5% discount bracket is the **Revenue King (37,005 USD)**. **Recommendation:** Use a "teaser" 5% discount in the UK to maximize profit. Avoid the 30% "Profit Killer" tier, as it results in the lowest revenue efficiency in both countries.
*   **Brand-Specific Discounting:**
    *   **Nike & New Balance:** Maintain full price. They are "Discount Resistant" (low sensitivity).
    *   **Reebok & Adidas:** Can use tactical 10-15% discounts in the USA to clear inventory, as they show higher sensitivity (0.22 and 0.14).

### **2. Regional Market Mastery (UK vs. USA)**
*   **Targeting by Income:**
    *   **UK:** Focus marketing on **High-Income earners**, who drive 64k USD in revenue and show almost zero sensitivity to price cuts.
    *   **USA:** Focus on **Medium-Income earners** (63k USD revenue), while using targeted discounts only for the **Low-Income** segment, which is the only US group that truly responds to sales (0.15 sensitivity).
*   **The Satisfaction Gap:** UK customers are more satisfied (4.0) than US customers (3.9). In the US, high-priced shoes (175 USD+) saw a **downward trend in ratings for 2025**, indicating a "Value Deficit." **Recommendation:** Conduct a quality-control audit on premium US lines to ensure the product justifies the price.

### **3. Product & Category Dominance**
*   **Leverage Brand Strengths:**
    *   **Nike:** Continue dominating the **Basketball** category (42k USD).
    *   **New Balance:** Capitalize on your position as the **2025 Volume Leader (1,510 units)**. You are winning on brand desire, especially in the Gym and Training categories.
    *   **Adidas:** Double down on **Lifestyle and Running**, where you currently hold the market lead.
*   **The Unisex Opportunity:** Unisex footwear is the fastest-growing volume category in the UK. **Recommendation:** Pivot design and marketing toward gender-neutral styles for the 2026 UK collection to capture this rising momentum.

### **4. 2026 Predictive Outlook**
*   **The "Adidas USA" Boom:** Predictively, Adidas has the highest growth momentum in the USA. They are projected to reach **38,280 USD** in 2026 revenue, successfully overtaking Nike. **Action:** Invest heavily in US logistics and marketing for Adidas.
*   **The "Nike UK" Stability:** Nike is the steady "Momentum King" of the UK, projected to reach **38,561 USD**. **Action:** Use Nike as your stable "cash cow" in Europe.
*   **Reject the 10% Blanket Discount:** Our simulation confirms that a 10% discount across the board in H1 2026 would result in a **net loss of nearly 10,000 USD for Nike**. **Action:** Pivot from "Price Wars" to "Value-Added" marketing (e.g., exclusive drops, loyalty rewards).
